```python
#Outliers and Anomalies
# Ensure you have pivot_df created as before
pivot_df = df.pivot_table(
    index=['area', 'item', 'year'],
    columns='element',
    values='value',
    aggfunc='mean'
).reset_index()
```

```python
#. Detect Outliers Using IQR Method
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] < lower) | (df[column] > upper)]

# Detect outliers
yield_outliers = detect_outliers_iqr(pivot_df, 'Yield')
production_outliers = detect_outliers_iqr(pivot_df, 'Production')
```

```python
# Visualize Anomalies Over Time
import matplotlib.pyplot as plt
import seaborn as sns

# Yield over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_df, x='year', y='Yield', label='All Data')
sns.scatterplot(data=yield_outliers, x='year', y='Yield', color='red', label='Yield Outliers')
plt.title('Yield Over Years with Outliers Highlighted')
plt.tight_layout()
plt.show()

# Production over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_df, x='year', y='Production', label='All Data')
sns.scatterplot(data=production_outliers, x='year', y='Production', color='red', label='Production Outliers')
plt.title('Production Over Years with Outliers Highlighted')
plt.tight_layout()
plt.show()
```

```python
# Count outliers by crop and region
print("Top Crops with Yield Outliers:")
print(yield_outliers['item'].value_counts().head())

print("\nTop Regions with Production Outliers:")
print(production_outliers['area'].value_counts().head())

print("\nOutlier Years (Yield):", yield_outliers['year'].value_counts().sort_index())
print("Outlier Years (Production):", production_outliers['year'].value_counts().sort_index())
```
